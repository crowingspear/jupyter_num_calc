## よく使うkeyboard short cut
1. 入力はshift+enter
1. 改行はenter

1. command mode(cellが青色)
  1. hでhelpを表示
  1. escでedit modeへ
  1. ^jは下にcellを挿入
  1. ^kは上にcellを挿入
  1. ^xでcell cut
  1. ^yでcell paste
  1. ^mはcellをmarkdown cellに変更
  1. yでcode cellに変更
1. edit mode(cellが緑色)
  1. emacsのkeybindが使える．うまくいかないときは，closeして再openするといい．
  1. escでcommand modeへ

# 代数式

例えば，
$$
(x+y)^2 = x^2 + 2xy + y^2
$$
という展開をおこなったり，微分，積分，線形代数計算を$x,y, A, b$
なんかのシンボルでそのまま変形したい時があります．

こういうことをおこなってくれるシステムが数式処理システム(Computer algebra systems)と呼ばれるものです．
代表的な商用製品としては，Maple, Mathematicaがあります．
同じコンセプトでフリーのものには，Octave, Maxima, SciLabなどがあります．

これをpython上で実現しようとしたのが，sympy, sageです．
sageのほうが入出力が綺麗なんですが，残念ながらwindowsでは動きません．
したがって，sympyでしばらくどこまでできるかを試してみます．


```python
from sympy import Symbol
x = Symbol('x')
print(x + x + 1)
```

```python
a = Symbol('x')
print(a+a+1)
```

```python
print(x.name)
print(a.name)
```

```python
from sympy import symbols
x,y,z = symbols('x,y,z')
s = x*y + y*x
print(s)
```

```python
p = x*(x+x)
print(p)
```

```python
from sympy import pprint
pprint(p)
```

```python
pprint(p.integrate(x))
```

# plot

簡単なplotです．

1. 複数の関数のplot, 色の変更まで．
1. parametric plotして円をaspect_ratio 1:1で．
1. sin(x) * cos(y) ぐりぐりはできなさそう．

```python
from sympy import Symbol
from sympy.plotting import plot
import sympy

x=Symbol('x')
plot(sympy.sin(x))
```

```python
p = plot(sympy.sin(x),sympy.cos(x),
         legend=True, show=False)
p[0].line_color = 'b'
p[1].line_color = 'r'
p.show()
```

```python
from sympy import symbols, cos, sin, pi
from sympy.plotting import plot_parametric
t = symbols('t')

import matplotlib.pyplot as plt
plt.rcParams['figure.figsize'] = (8,8)

plot_parametric(cos(t), sin(t), (t, 0, 2*pi))
```

```python
from sympy import symbols, sin, cos, pi
from sympy.plotting import plot3d
x, y = symbols('x y')

plot3d(sin(x)*cos(y), (x,-pi,pi),(y,-pi,pi))
```
